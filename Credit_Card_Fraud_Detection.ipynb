Splitting the Dataset

```python
X = data.drop('Class', axis=1)
y = data['Class']


#Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
#Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#SMOTE to balance the dataset
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)
```

Training the XGBoost Classifier

```python
xgb_classifier = XGBClassifier(scale_pos_weight=10, n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
xgb_classifier.fit(X_train_smote, y_train_smote)
```

Model Evaluation

```python
y_probs = xgb_classifier.predict_proba(X_test_scaled)[:, 1]
y_pred_adjusted = (y_probs >= 0.5).astype(int)
```

Evaluating the Model with Adjusted Threshold

```python
accuracy_adjusted = accuracy_score(y_test, y_pred_adjusted)
conf_matrix_adjusted = confusion_matrix(y_test, y_pred_adjusted)
class_report_adjusted = classification_report(y_test, y_pred_adjusted)

print("Accuracy with adjusted threshold:", accuracy_adjusted)
print("Confusion Matrix with adjusted threshold:\n", conf_matrix_adjusted)
print("Classification Report with adjusted threshold:\n", class_report_adjusted)
```

Precision-Recall AUC

```python

precision, recall, _ = precision_recall_curve(y_test, y_probs)
auc_pr = auc(recall, precision)
print("AUC-PR:", auc_pr)
```
